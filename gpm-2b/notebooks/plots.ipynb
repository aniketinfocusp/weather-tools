```python
def get_all_dataset_from_group(group_name, file):
    """Returns all the datasets available under """
    group = file[group_name]
    paths = []
    for item in group.items():
        if(type(item[1])==h5py._hl.dataset.Dataset):
            paths.append(f"{group_name}/{item[0]}")
    return paths
```

```python
def plotter(lat, lon, weather_data, title, cmap="rainbow", large=False):
    """Simple data plotter
        - cmap can be passed for better visualization
        - Size of plot can be changed by passing large=True    
    
    """
    figsize = (10, 5)
    if(large):
        figsize = (20, 20)
    plt.subplots(figsize=figsize)

    my_map = Basemap(projection='mill', lon_0=0)
    my_map.drawcoastlines()
    x, y = my_map(lon, lat)

    vmax = np.max(weather_data)
    vmin = np.min(weather_data)
    
    im = my_map.scatter(x, y, c=weather_data, cmap=cmap, zorder=1, s=1)
    plt.title(title)
    plt.show()
```
